```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
###draw a line in a diagram from position (1,2) to position (6,7):
x = np.array([1,2])
y = np.array([6,7])

plt.plot(x,y)
plt.show()
```

```python
x=np.array([1,10])
y=np.array([10,20])

plt.plot(x,y)
plt.show()
```

```python
x=np.array([1,10])  ##'*' is wsed if we want only * insted of line 
y=np.array([10,20])

plt.plot(x,y,'*')
plt.show()
```

```python
y=np.array([10,20,30,40])  ## if we want colour of the marker, if we won give y value it ll take defult values,: for doted lines
plt.plot(y,'*'':r')
plt.show()
```

```python
y=np.array([10,20,30,40])  ## if we want colour of the marker, if we won give y value it ll take defult values
plt.plot(y,'*'':b')
plt.show()
```

```python
x=np.array([1,2,3,4])
y=np.array([3,6,5,8])

plt.plot(x,y,color='r',linewidth=2)

plt.xlabel("x-axies",fontsize=10,color='green')
plt.ylabel("y-axies",fontsize=10,color='green')

plt.show()
```

```python
x=np.array([1,2,3,4])
y=np.array([3,6,5,8])

plt.plot(x,y,color='r',linewidth=2)

plt.title("school data",loc="left",fontsize=30,color='blue')
plt.xlabel("x-axies",loc='left')
plt.ylabel("y-axies")

plt.plot(x,y)
plt.show()
```

```python
x=np.array([1,2,3,4])
y=np.array([3,6,5,8])

plt.plot(x,y,color='r',linewidth=2)

plt.title("school data")
plt.xlabel("x-axies")
plt.ylabel("y-axies")

plt.plot(x,y)
plt.show()
```

```python
x=np.array([1,2,3,4])
y=np.array([3,6,5,8])

plt.plot(x,y,color='r',linewidth=2)

plt.title("school data")
plt.xlabel("x-axies")
plt.ylabel("y-axies")

plt.grid()

plt.plot(x,y)
plt.show()
```

```python
x=np.array([1,2,3,4])
y=np.array([3,6,5,8])

plt.plot(x,y,color='r',linewidth=10)

plt.title("school data")
plt.xlabel("x-axies")
plt.ylabel("y-axies")

plt.grid(color='green',linestyle='--',linewidth=5)

plt.plot(x,y)
plt.show()
```

```python
x=np.array([1,2,3,4])
y=np.array([3,6,5,8])

plt.plot(x,y,color='r',linewidth=2)

plt.title("school data")
plt.xlabel("x-axies")
plt.ylabel("y-axies")

plt.grid(color='green',linestyle='--',linewidth=2.3)

plt.plot(x,y)
plt.show()
```

```python
x=np.array([10,20,30,40])
y=np.array([30,40,50,60])

colors=(['red','blue','green','black'])
size=([10,20,30,10])
plt.scatter(x,y,c=colors,s=size)

plt.show()
```

```python
x=np.array(["a","b","c"])
y=np.array([1,2,3,])

plt.bar(x,y)
plt.show()
```

```python
x=np.array(["a","b","c"])##barh for horizatal display
y=np.array([1,2,3,])

plt.barh(x,y)
plt.show()
```

```python
x=np.array(["a","b","c"])
y=np.array([1,2,3,])

plt.bar(x,y,color="red")
plt.show()
```

```python
x=np.array(["a","b","c"])
y=np.array([1,2,3,])

plt.bar(x,y,width=0.1)
plt.show()
```

```python
y=np.array([10,20,30,40])

plt.pie(y)
plt.show()
```

```python
y=np.array([10,20,30,40])

mylabels=["a","b","c","d"]
plt.pie(y,labels=mylabels)
plt.show()
```

```python
y=np.array([10,20,30,40])

mylabels=["a","b","c","d"]
mycolors=["red","pink","green","orange"]
plt.pie(y,labels=mylabels,colors=mycolors)
plt.show()
```

```python
y=np.array([10,20,30,40])

mylabels=["a","b","c","d"]
mycolors=["red","pink","green","orange"]
myexplode=[0,0.5,0.1,1]

plt.pie(y,labels=mylabels,colors=mycolors,explode=myexplode)
plt.show()
```

```python
y=np.array([10,20,30,40])
mylabels=["a","b","c","d"]
plt.pie(y)
plt.legend(labels=mylabels)
plt.show()
```

```python
y=np.array([10,20,30,40])

mylabels=["a","b","c","d"]
plt.pie(y,labels=mylabels,startangle=180)
plt.show()
```

```python
y=np.array([10,20,30,40])

mylabels=["a","b","c","d"]
plt.pie(y,labels=mylabels,startangle=90)
plt.show()
```
